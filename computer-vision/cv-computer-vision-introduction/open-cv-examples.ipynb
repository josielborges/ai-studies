# Computer Vision Introduction

### Print the image

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("face.png")

# cv2.imshow("Image", image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_image)
plt.axis("off")
plt.show()
```

### Print the image in gray scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.show()
```

### Resize the image

```python
width = 800
height = 600

sizes = (width, height)

resized_image = cv2.resize(image, sizes, interpolation=cv2.INTER_AREA)

rgb_resized_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_resized_image)
plt.axis("off")
plt.show()
```

### Print the image with soft edges

```python
soft_image = cv2.GaussianBlur(image, (15, 15), 0)

rgb_soft_image = cv2.cvtColor(soft_image, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_soft_image)
plt.axis("off")
plt.show()
```

### Print the image with edges

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

borders = cv2.Canny(gray_image, 100, 200)

plt.imshow(borders, cmap="gray")
plt.axis("off")
plt.show()
```

### Print the image with rectangle

```python
start = (50, 5)
end = (620, 560)
color = (255, 0, 0)
border_size = 2

image_with_rectangle = cv2.rectangle(image.copy(), start, end, color, border_size)

rgb_imarge_with_rectangle = cv2.cvtColor(image_with_rectangle, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_imarge_with_rectangle)
plt.axis("off")
plt.show()
```

```python
output_path = "image_with_rectangle.png"
cv2.imwrite(output_path, image_with_rectangle)
```
